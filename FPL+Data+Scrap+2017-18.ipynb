```python
# Importing the dependant Libraries and setting system encoding
import requests
import pandas as pd
```

```python
def dict_to_df(d):
    df=pd.DataFrame(d.items())
    df.set_index(0, inplace=True)
    return df.T
# Getting Player info
url = 'https://fantasy.premierleague.com/drf/bootstrap-static'
r = requests.get(url)
rawglobal = r.json()
playerinfo = rawglobal['elements']
dfplaya = pd.DataFrame()
for info in playerinfo:
    playerdict = info
    dfcurr = dict_to_df(playerdict)
    dfplaya = pd.concat([dfplaya, dfcurr], axis=0)
```

# Update the below two variables for Each gameweek

```python
Gameweek = 1
maxplayers = 510
```

```python
# Getting Playerdata
df = pd.DataFrame()
for i in range(maxplayers):
    plyerid = i+1
    url = 'https://fantasy.premierleague.com/drf/element-summary/' + str(plyerid)
    r = requests.get(url)
    rawdata = r.json()
    playerdata = rawdata['history']
    for fixture in playerdata:
        playerdict = fixture
        playerdict['player_id'] = i+1
        dfcurr = dict_to_df(playerdict)
        df = pd.concat([df, dfcurr], axis=0)
    print 'Player id ' + str(i+1)
```
